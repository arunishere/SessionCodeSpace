### Data Analysis with Pandas - III

```python
'''

1. Histogram - Represent the Distribution of the Data ; Continous Data
2. Bar Chart - Represent Frequency
3. Pie Chart -  Represents Share
4. Line Chart - Time or Change over time
5. Box Plot - Distribution & Spead of the Data 
6. Scatter Plot - Relationship between two Variable

'''
```

#### Importing Modules & Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('OrangeQ.csv')
```

```python
df.head()
```

```python
df.columns
```

#### Histogram & Bar Charts

```python
sns.histplot(data=df, x='Weight(g)')
plt.title('Weight Distribution of Oranges')
```

```python
sns.histplot(data=df, x='Size(cm)')
plt.title('Size Distribution of Oranges')
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
sns.histplot(data=df, x='Weight(g)')
plt.title('Weights')

plt.subplot(1,2,2)
sns.histplot(data=df, x='Size(cm)')
plt.title('Sizes')

plt.suptitle('Distribution')
plt.show()
```

```python
df['Color'].value_counts().plot.bar()
plt.xticks(rotation=45)
plt.title('Color Frequency')
plt.xlabel('Orange Color')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Color'].value_counts().plot.barh()
```

```python
sns.countplot(df['Color'])
```

#### Box & Scatter Plots

```python
sns.boxplot(df['Size(cm)'])
```

```python
sns.boxplot(df['Weight(g)'])
```

```python
df.columns
```

```python
plt.figure(figsize=(10, 6))

sns.boxplot(df[['Size(cm)','Brix(Sweetness)', 'pH(Acidity)',
       'Softness(1-5)', 'HarvestTime(days)', 'Ripeness(1-5)', 'Color',
       'Variety', 'Blemishes(Y/N)']])

```

```python
sns.scatterplot(data=df, y='Size(cm)', x='Weight(g)')
```

```python

plt.title('Co-relation b/w Harvest Time & Size')
sns.scatterplot(data=df, y='Weight(g)', x='HarvestTime(days)', hue='Color',legend=False, s=50, marker="*")
plt.show()
```

#### Line Chart

```python
dates = pd.date_range('2024-01-01', '2024-06-30')
```

```python
value = np.random.randint(20000,21500,size=(182))
values = value + np.random.rand(182)
```

```python
plt.figure(figsize=(18, 6))
plt.plot(dates, values)
plt.title('Trading Trend')
plt.ylabel('Value ()')
plt.xlabel('Trading Days')
plt.plot(dates, np.array([20733.556191228152 for i in range(182)]))
plt.legend(['Normal Trend', 'Baseline'])
plt.show()
```

#### Pie Chart

```python
plt.title(r'% Share of Color of Oranges')
plt.pie(df['Color'].value_counts(), autopct=r'%1.1f%%', shadow=True)
plt.show()
```
